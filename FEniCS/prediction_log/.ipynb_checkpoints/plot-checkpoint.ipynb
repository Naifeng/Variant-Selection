```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(15,8))

sns.set(style="ticks", font_scale=1.45)
df = pd.read_csv("Blur_2^15_data.csv")
ax = sns.barplot(x="schedule", y="runtime", data=df, hue='shape', dodge=False)
# ax = sns.scatterplot(x="schedule", y="runtime", data=df, s=50)
#ax.set_title('Blur_3x3 with 2^15 input')
ax.set_xlabel('schedule')
ax.set_ylabel('execution time (s)')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[0:], labels=labels[0:])

for line in range(0,df.shape[0]):
     ax.text(line+0.01, df.runtime[line]+0.05, df.runtime[line], horizontalalignment='center', 
     size='medium', color='black')

#ax.text(-0.2, 5.5, '3.80', size='medium', color='black', )
#ax.text(0.9, 8.5, '6.60', size='medium', color='black')
#ax.text(1.9, 4.5, '2.63', size='medium', color='black')
#ax.text(2.8, 46.5, '50.98', size='medium', color='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
import pandas as pd
import numpy as np


rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)


plt.figure(figsize=(15,8))

sns.set(style="ticks", font_scale=3.5, font='text.usetex', palette='Dark2')

df = pd.read_csv("result_plot.csv")

ax = sns.lineplot(x="mesh_size", y="runtime", hue="shape", 
                  style='shape',
                  markers=True, data=df, linewidth=5, markersize=15)
#ax.set_title('Blur_3x3 with 2^15 input')


ax.lines[0].set_linestyle("-")
ax.lines[1].set_linestyle("None")

ax.lines[1].set_marker("^")

ax.set_xlabel('input mesh size')
ax.set_ylabel('execution time (s)')

ax.set(yscale="log")
ax.set_ylim([1e-2, 100])



#ax.legend().get_lines()[3].set_linestyle("None")

handles, labels = ax.get_legend_handles_labels()
handles[2].set_linestyle("None")

handles[1].set_markersize(15)
handles[2].set_markersize(15)


handles[2].set_marker("^")

handles[1].set_linewidth(3)
handles[2].set_linewidth(3)


handles[1].set_linestyle("-")
#handles[2].set_linestyle("--")



ax.legend(handles=handles[1:], labels=labels[1:])



#for line in range(0,df.shape[0]):
#     ax.text(line+0.01, df.runtime[line]+0.05, df.runtime[line], horizontalalignment='center', 
#     size='medium', color='black')



plt.show()
```
